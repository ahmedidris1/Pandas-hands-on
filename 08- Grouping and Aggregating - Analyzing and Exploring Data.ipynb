# Data Aggregation

### Aggregate Functions

They are functions that return a value which is made up of proccessing several other values, like `mean()`, `median()` and `mode()`.       

```python
import pandas as pd
```

```python
survey_df = pd.read_csv('data/survey_results_public.csv')
schema_df = pd.read_csv('data/survey_results_schema.csv')
```

### An aggregate function applied on a `Series`     
To know a typical average salary of the developers in StackOverflow dataset (`survey_df`), we can use the `median()` method on the column of the developers salaries which is named `ConvertedComp`. We use `median()` instead of `mean()`, because `median()` is less affected by outliers.   

```python
survey_df['ConvertedComp']
```

```python
survey_df['ConvertedComp'].median()
```

### An aggregate function applied on a `DataFrame`      
Now, instead of applying an aggregate function on a `Series`, let's see the result of applying an aggregate function on a `DataFrame`.

```python
survey_df.median()
```

`median()` method will automatically operate only on columns `Series`es with numerical values and ignore the columns with none numerical values.

### Brief statistical summary aggregate function `describe()`      
There is a function that when is applied returns a brief statistical summary called `describe()`.     
it returns a result with the following parameters:-      
        __count:__ The number of truthy values in a column.       
         __mean:__ The average.    
          __std:__ The standard deviation.    
          __min:__ The minimum value.     
__25%, 50%, 75%:__ The first quantile, the median and the third quantile.     
          __max:__ The maximum value.

```python
survey_df.describe()
```

### Apply `describe()` on a `Series`     
`describe()` is applicable on `Series` too.            
       
__example:__       
Let's consider having a statiscal brief about the salary column `ConvertedComp`.

```python
survey_df['ConvertedComp'].describe()
```

__Note:__ `count` value at the top of the result represents the number of valid entries in `ConvertedComp` column. Therefore any `None` value or none numerical value will get ignored.     
`count()` is different from `value_counts()` which tallies the frequency of each unique entry in a `Series`.

```python
survey_df['SocialMedia']
```

Now we can see a the popularity of social apps among all developers, but __what if we want to check the popularity of social media apps in each country?__. The basic answer for this is to use filters as we have learned in filtering data notebook.     
__example:__      
Let's consider checking the popularity of social media apps in `"United States"`.
we are going to make a filter  `_filter` = `survey_df["Country"]` == "United States"      
`survey_df.loc[_filter, 'SocialMedia']`

```python
 _filter = survey_df["Country"] == "United States"
```

```python
survey_df.loc[_filter, 'SocialMedia']
```

```python
survey_df.loc[_filter, 'SocialMedia'].value_counts()
```

Now if we want to get the popularity of social media apps in another country, we have to filter data according to the desired country. It is clear that this approach requires repeating steps and is tedious to use, what if we can get the popularity of social media apps in each country, all in one step. Here the method `groupby()` comes in handy.

```python
help(pd.DataFrame.groupby)
```

`groupby()` takes as an argument a __list of columns__ to group by. Therefore, if we want to group our data by `'Country'` column we have to pass `'Country'` as an element in a list of columns to group by.

```python
country_grp = survey_df.groupby(['Country'])
country_grp
```

As we see, the returned value of `.groupby()` method is a special `Pandas` object that contains several groups inside it.     

If we ran `.value_counts()` on `country_grp` object, we would get a `multiple indexed Series`

```python
country_grp['SocialMedia'].value_counts().head(50)
```

### Get a specific group out of a Pandas `groupby` object     
To get a specific group, use `.get_group()` method on a `groupby` object.

```python
country_grp.get_group("United States")
```

```python
country_grp.get_group("United States")["SocialMedia"].value_counts()
```

if we want to check social media apps popularity for another country we only need to pass the desired country name in     
`country_grp.get_group("CountryName")["SocialMedia"].value_counts()`, unlike the filter approach which requires creating a new filter for every country.

```python
country_grp.get_group("India")["SocialMedia"].value_counts()
```

Let's get the `median` of salaries per country. We select the column `ConvertedComp` and apply the aggregate function `.median()`.

```python
country_grp['ConvertedComp'].median()
```

if we want to get the median salary for a specific country, 'Germany' for example, we can access it directly from the result above by passing 'Germany as an index for `country_grp['ConvertedComp'].median()['Germany']` .

```python
country_grp['ConvertedComp'].median()['Germany']
```

__OR__      
use the get_group on country_grp first, then select the salaries column `convertedComp` and then apply the `median()` method. 

```python
country_grp.get_group('Germany')['ConvertedComp']
```

```python
country_grp.get_group('Germany')['ConvertedComp'].median()
```

__OR__       
Apply the `median()` aggregate method on `country_grp` `groupby` object and then access 'Germany' median value using `.loc` function.

```python
country_grp['ConvertedComp'].median()
```

```python
country_grp['ConvertedComp'].median().loc['Germany']
```

### Apply Several aggregate functions all at once      
If we want to run several aggregate functions, we use `agg` method which takes as an argument a __list of the aggregate functions__ that we want to apply.     
__example:__      
If we want to apply `median()` and `mean()` aggregate functions on `country_grp['ConvertedComp']`, we do the following:    
`country_grp['ConvertedComp'].agg(['median', 'mean'])`.

```python
country_grp['ConvertedComp'].agg(['median', 'mean'])
```

Now we can get the mean and median for any country in the result above just by accessing it with its name as an index.    
__example:__     
Get the mean and the median of salaries in 'Canada'.     
`country_grp['ConvertedComp'].agg(['median', 'mean']).loc['Canada']`.

```python
country_grp['ConvertedComp'].agg(['median', 'mean']).loc['Canada']
```

### Tricky example    
If we want to know the number of individuals who know Python in each country, we can use the filter approach to investigate each country separately.

__1<sup>st</sup> approach:__     
    Make a filter that targets individuals from the desired country, let it be india for example, then access the column `'LanguageWorkedWith'` which contains the data of languages that each individual is familiar with. Then apply the `pd.str.contains(arg)` method that check if __`arg`__ is included or not. Finally, we count how many times __'Python'__ is included in the boolean values that resulted from `.str.contains('Python')`, using `sum()`.

```python
_filter = survey_df['Country'] == 'India'
```

```python
survey_df[_filter]['LanguageWorkedWith']
```

```python
survey_df[_filter]['LanguageWorkedWith'].str.contains('Python')
```

```python
survey_df[_filter]['LanguageWorkedWith'].str.contains('Python').sum()
```

__OR__      
__An alternative way for the 1<sup>st</sup> approach__

```python
_filter = _filter & survey_df['LanguageWorkedWith'].str.contains('Python')
```

```python
rows, cols = survey_df[_filter].axes
```

```python
len(rows)
```

__OR__

```python
len(survey_df[_filter].axes[0])
```

__2<sup>nd</sup> approach:__     
__Use groupby() method__

if we want to use `groupby()` method, we might primarly think of something like the following:-     
`survey_df.groupby(['Country'])['LanguageWorkedWith'].str.contains('Python')`.

```python
survey_df.groupby(['Country'])['LanguageWorkedWith']
```

```python
survey_df.groupby(['Country'])['LanguageWorkedWith'].str.contains('Python')
```

Unfortunately, the `groupby` object does not provide a direct access to use `str` class methods, we have to use `.apply()` method instead.

```python
survey_df.groupby(['Country'])['LanguageWorkedWith'].apply(lambda grp: grp.str.contains('Python').sum())
```

To get the percentage of the individuals who know Python per country.

```python
knows_python_pct = survey_df.groupby(['Country']).apply(lambda grp: (grp['LanguageWorkedWith'].str.contains('Python').sum() / len(grp.axes[0]) ) * 100 )
```

```python
knows_python_pct['United States']
```

__OR__     
We can build a new `DataFrame` that describes the the data related to individuals who know Python per country.      
__The first step__ is to know the number of individuals per each country, we will call it `country_respondants`.     
__The second step__ is to know the number of individuals who know Python per each country, we will call it `country_uses_python`.

```python
country_respondants = survey_df['Country'].value_counts()
country_respondants
```

```python
country_uses_python = survey_df.groupby(['Country'])['LanguageWorkedWith'].apply(lambda grp: grp.str.contains('Python').sum())
country_uses_python
```

The next step is to concatenate `country_respondants` and `country_uses_python` in one `DataFrame`. we achieve this step by using `pd.concat()` method.      
`pd.concat()` takes as an argument a list of the columns that are going to be concatenated together, in addition to setting `axis` parameter to `columns/rows` (`rows` is the default value), and `sort` parameter has to be set to `False`

```python
python_df = pd.concat([country_respondants, country_uses_python], axis='columns', sort=False)
python_df
```

Let's rename the columns names, `Country` -> `NumRespondants`, `LanguageWorkedWith` -> `NumKnowsPython`.

```python
python_df.rename(columns={'Country': 'NumRespondants' , 'LanguageWorkedWith' : 'NumKnowsPython'})
```

```python
python_df.rename(columns={'Country': 'NumRespondants' , 'LanguageWorkedWith' : 'NumKnowsPython'}, inplace=True)
```

Now we are going to add a new column that represents the percentage of the people who know Python in each country.

```python
python_df['PctKnowPython'] = (python_df['NumKnowsPython']/python_df['NumRespondants']) * 100
```

```python
python_df
```

Let's sort `python_df` by `PctKnowPython` descendingly.

```python
python_df.sort_values(by='PctKnowPython', ascending=False)
```
