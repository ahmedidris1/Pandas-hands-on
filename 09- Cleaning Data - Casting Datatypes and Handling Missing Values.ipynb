# Handling Missing Values         
There are several actions for handling missing data, one of them is to remove it. 

```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    "last": ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    "email": ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    "age": ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

As we see in the `DataFrame` above __df__ that there are standard null values in a shape of `None`, `np.nan` beside a custom `String` missing values like `Missing` and `NA`.        
Generally when we want to drop missing values, we use the function `DataFrame.dropna()` 

```python
df.dropna()
```

As we see, `DataFrame.dropna()` drops any row that contains a standard null value (like `None`, `np.nan`). Therefore, the rows 3, 4, 5 have been removed. `DataFrame.dropna()` has parameters that are set to default values by default, by changing these parameters' values, the behaviour of `DataFrame.dropna()` gets changed.        
   
   
`DataFrame.dropna()` has the following parameters:        

`axis` = "`index`" => defines the axis to drop missing data from, `axis` has two possible value options, __index__/__columns__, __index__ means __rows__.       
       
`how` = "`any`" => defines the criteria which based on it, data get dropped, `how` has two possible value options, __any__/__all__, "__`any`__" means, if any data in the selected `axis` is missing, drop the selected `axis`, whilist __all__ mean if all data in the selected `axis` are missing then drop the selected `axis`.

```python
df.dropna(axis = "index", how = "any")
```

`df.dropna(axis = "index", how = "any")` gets us the same result as using `df.dropna()`, because __axis__ = "index", __how__ = "any", are the default values.

```python
df.dropna(axis = "index", how = "all")
```

`df.dropna(axis = "index", how = "all")` drops rows when all cell in a row have missing values. The only row that has all its cell with missing values was the row 4.

```python
df.dropna(axis = "columns", how = "all")
```

`df.dropna(axis = "columns", how = "all")` will not remove any column, because there is no column that has its all cells with missing values.

```python
df.dropna(axis = "columns", how = "any")
```

`df.dropna(axis = "columns", how = "any")` gets us an empty result, because the row number 4 has all its cell with missing values, therefore, each column has at least one missing value, and because dropping criterea __`how`__ = "`any`", any column with even an only missing value will get dropped.

### drop rows when there are missing data based on specific columns not all cells.         
To do so, use the parameter __`subset`__ and assign to it a list of the columns that you want to watch whether it contain missing values.     
__example:__       
consider that we want to drop rows if a row has no email only, no problem if there is missing `first`, `last` or `age`.

```python
df.dropna(axis = "index", how = "any", subset = ['email'])
```

`df.dropna(axis = "index", how = "any", subset = ['email'])` has dropped `row 3` and `row 4` as their `email` column contain `standard nullish values`, only left the `row 6` which contains a custom `nullish value` at `email` cell.

When __subset__ list as an argument in `DataFrame.dropna()`, contains only one column value, it make __how__ argument with no use. because there is only one value to decide upon dropping a row or not, __all__/__any__ will have no effect as they need multiple columns to be of a use.

```python
df.dropna(axis = "index", how = "all", subset = ['last', 'email'])
```

Only the 4th column has been dropped as `last` and `email` columns are both with `Nullish` values.

### Custom missing values        
Sometimes missing data are entered in a string format (like `"Missing"` and `"NA"`) not in standard formats like `None`, `np.nan`. The traditional way to replace such value, which is by spotting the custom values for missing data first and then manually replace them by using:     
`DataFrame.replace(old_value, new_value, inplace=True)`.        
      
__*Note:__       
When `DataFrame` is constructed by `pd.read_csv` there is a way to clean data by passing some arguments to `pd.read_csv`. 

```python
# change "Missing" to NaN
df.replace("Missing", np.nan, inplace=True)
```

```python
# change "NA" to NaN
df.replace("NA", np.nan, inplace=True)
```

```python
df
```

### Check which values in a `DataFrame` considered nullish       
To check the status of each value in a `DataFrame`, wether it is considered `Nullish` value or not, run the function `DataFrame.isna()`.

```python
df.isna()
```

__True__ => Nulish

### Fill cells which have Nullish values with a default value        
To fill the cells that contain Nullish values with a specific value, use the function DataFrame.fillna(filling_value, inplace=True).

```python
df.fillna('Missing')
```

### Casting the type of data       
The column age appears as column of integers, but if we returned back to the `people` dictionary we will find that the `age` key has a valuse that is a list of ages in a shape of Strings. we can check that by calling the property `DataFrame.dtypes` on the `DataFrame` `df`, in order to check what is the type of each column.

```python
df.dtypes
```

__object__ => String or a mix of types.      
and therefore we can not perform mathematical operations on the column `age`.

```python
df['age'].mean()
```

### Note:-       
When casting the type of a column's data that contains `NaN` values, never cast it to integer but cast it to float, as `NaN` is a float value under the hood. 

```python
type(np.nan)
```

### Casting a column's data type syntax:       
to cast a column's data type syntax. DataFrame['col_name] = DataFrame['col_name].astype(target_type_of_casting) 

```python
# casting the age column into float
df['age'] = df['age'].astype(float)
```

```python
df.dtypes
```

```python
df['age'].mean()
```
